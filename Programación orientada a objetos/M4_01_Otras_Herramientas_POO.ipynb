**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
class Vehiculos: # inicializamos constructor de vehiculos
    def __init__(self, marca, modelo):
        self.__marca = marca
        self.__modelo = modelo

    @property 
    def marca(self):
        return self.__marca

    marca.setter
    def marca(self, nueva_marca):
        self.__marca = nueva_marca

    @property 
    def modelo(self):
        return self.__modelo

    modelo.setter
    def marca(self, nueva_modelo):
        self.__modelo = nueva_modelo

    def repostar(self): # funcion repostar
        print("Este vehiculo tiene que repostar gasolina")

    def __str__(self):
        return "El vehiculo es {}, {}".format(self.__marca, self.__modelo)

class VElectricos(Vehiculos): # inicializo VElectricos
    def __init__(self, modelo, marca, autonomia):
        super().__init__(marca, modelo)
        self.__autonomia = autonomia

    @property 
    def autonomia(self):
        return self.__autonomia

    autonomia.setter
    def autonomia(self, nueva_autonomia):
        self.__autonomia = nueva_autonomia

    def repostar(self): # funcion repostar
        print("Este vehiculo tiene que repostar electricidad")
        
    def __str__(self):
        return super().__str__() + " y tiene una autonomia de {}.".format(self.__autonomia)

class BicicletaElectrica(VElectricos): 
    pass # VElectricos hereda de Vehiculo por lo que he creido conveniente hacerlo asi, es herencia simple multinivel, mas simple, menos errores

class Quad(Vehiculos): # inicializo quad, he decidido que no haya herencia multiple porque me da error, porque VElectrico es 
    def __init__(self, marca, modelo, electrico:VElectricos = None): #hijo de vehiculo
        super().__init__(marca, modelo)
        # self.__autonomia = autonomia
        self.__electrico = electrico # he creado un atributo nuevo que dice si es electrico o no, llamacdo a VElectricos

    @property  
    def electrico(self):
        return self.__electrico

    @electrico.setter
    def electrico(self, nuevo_electrico):
        self.__electrico = nuevo_electrico

    def repostar(self):
        if self.__electrico: # si es electrico llama a la funcion de VElectricos
            self.electrico.repostar()
        else: # y sino llama a repostar de vehiculos
            super().repostar()

    def __str__(self):
        return super().__str__() + " {} es electrico".format(self.__electrico)
        
q1 = Quad(marca = "renault", modelo = "golf", electrico = VElectricos("Q", "Tgva", 200))
b1 = BicicletaElectrica(marca = "bh", modelo = "sd", autonomia = 180)
ve1 = VElectricos(marca = "2wed", modelo = "ff", autonomia = 180)
v1 = Vehiculos(marca = "bmw", modelo = "m3")
print(q1,"\n", b1,"\n", ve1,"\n", v1)
q1.repostar()
b1.repostar()
ve1.repostar()
v1.repostar()
```

**2) Ejercicio de herencia**
* Crea la superclase Personal_Universitario con un único atributo que sea un diccionario que incluya: "id", "nombre" y "email"
* Crear la clases Oficina, Profesor y Alumno. Todas ellas heredaran de Personal_Universitario y heredaran sus datos
* En el caso de Oficina, debe añadir a su diccionario de datos el dato "Puesto"
* En el caso de Profesor, debe añadir a su diccionario de datos el dato "Especialización"
* En el caso de Alumno, debe añadir a su diccionario de datos el dato "CreditosAprobados" (integer)
* Crear métodos que sirvan para mostrar la información de cada clase
* Crear objetos de todas las clases
* Mostrar la información de esos objetos y comprobar que sale lo que tiene que salir

```python
class Personal_Universitario:
    def __init__(self, dict): #se pasa el diccionario como parametro del constructor
        self.__id =  dict.get("id") # mediante get se añade, y el nombre del atributo seran las claves del dict
        self.__nombre = dict.get("nombre")
        self.__email = dict.get("email")

    @property 
    def id(self):
        return self.__id

    @id.setter
    def id(self, nuevo_id):
        self.__id = nuevo_id

    @property 
    def nombre(self):
        return self.__nombre

    @nombre.setter
    def nombre(self, nuevo_nombre):
        self.__nombre = nuevo_nombre

    @property 
    def email(self):
        return self.__email

    @email.setter
    def email(self, nuevo_email):
        self.__email = nuevo_email
    

    def __str__(self):
        return "El personal universitario: {}, tiene un id {}, y un email {}".format(self.nombre, self.id, self.email)

class Oficina(Personal_Universitario):
    def __init__(self, dict):
        super().__init__(dict)
        self.__puesto = dict.get("puesto")

    @property 
    def puesto(self):
        return self.__puesto

    @puesto.setter
    def puesto(self, nuevo_puesto):
        self.__puesto = nuevo_puesto   

    def __str__(self):
        return super().__str__() + "y su puesto es {}".format(self.puesto)

class Profesor(Personal_Universitario):
    def __init__(self, dict):
        super().__init__(dict)
        self.__especializacion = dict.get("especializacion") 

    @property 
    def especializacion(self):
        return self.__especializacion

    @especializacion.setter
    def especializacion(self, nueva_especializacion):
        self.__especializacion = nueva_especializacion

    def __str__(self):
        return super().__str__() + " y su especialización es: {}.".format(self.especializacion)
class Alumno(Personal_Universitario):
    def __init__(self, dict):
        super().__init__(dict)
        self.__creditos_aprovados = dict.get("creditos_aprovados")

    @property 
    def creditos_aprovados(self):
        return self.__creditos_aprovados

    @creditos_aprovados.setter
    def creditos_aprovados(self, nuevo_id):
        self.__creditos_aprovados = nuevo_id

    def __str__(self):
        return super().__str__() + "y tiene {} creditos aprovados".format(self.creditos_aprovados)   

pu1 = Personal_Universitario(dict = {"id" : 1, "nombre" : "pepe", "email" : "pepe@gmail.com"})
o1 = Oficina(dict = {"id" : 2, "nombre" : "pepa", "email" : "pepa@gmail.com", "puesto" : "administrativo"})
p1 = Profesor(dict = {"id" : 3, "nombre" : "pedro", "email" : "pedro@gmail.com", "especializacion" : "programacion"})
a1 = Alumno(dict = {"id" : 4, "nombre" : "pablo", "email" : "pablo@gmail.com", "creditos_aprovados" : 4})

print(pu1)
print(o1)
print(p1)
print(a1)

    
```

**3) Escoge la temática que prefieras y define una estructura de clases y subclases donde practiques la herencia en al menos dos niveles (herencia multi nivel). Crea una estructura de mínimo 5 clases, con sus respectivos constructores y métodos. Crea algunos objetos y comprueba que la herencia este funcionando. 
Si puedes, incluye alguna técnica de polimorfismo.**

```python
class Bebidas_alcoholicas: # clase abstracta
    def __init__(self, gaseosa, graduacion, color):
        self.__gaseosa = gaseosa
        self.__graduacion = graduacion
        self.__color = color

    @property 
    def gaseosa(self):
        return self.__gaseosa

    @gaseosa.setter
    def gaseosa(self, nueva_gaseosa):
        self.__gaseosa = nueva_gaseosa

    @property 
    def graduacion(self):
        return self.__graduacion

    @graduacion.setter
    def graduacion(self, nueva_graduacion):
        self.__graduacion = nueva_graduacion

    @property 
    def color(self):
        return self.__color

    @color.setter
    def color(self, nuevo_color):
        self.__color = nuevo_color

    def beber(self): # funcion hace el gesto de beber 
        print("La bebida alcoholica se esta bebiendo")

    def mezclar(self): # funcion que tras un input para preguntarle al usuario con que quiere mezclar la bebida, la mezcla
        mezcla = input("¿Con que refresco quieres mezclar la bebida?")
        print("Mezclando la bebida con {}".format(mezcla))

    def fermentar(self): # funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("Esta bebida no necesita fermentación.")

    def __str__(self):
        return "Esta es la clase abstracta, bebidas alcoholicas."

class Ginebra(Bebidas_alcoholicas):
    def __init__(self, gaseosa, graduacion, color, aromas, botanico_principal):
        super().__init__(gaseosa, graduacion, color)
        self.__aromas = aromas
        self.__botanico_principal = botanico_principal

    @property 
    def aromas(self):
        return self.__aromas

    @aromas.setter
    def aromas(self, nuevos_aromas):
        self.__aromas = nuevos_aromas

    @property 
    def botanico_principal(self):
        return self.__botanico_principal

    @botanico_principal.setter
    def botanico_principal(self, nuevo_botanico_principal):
        self.__botanico_principal = nuevo_botanico_principal

    def fermentar(self):  #funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("La ginebra se destila, no se fermenta.")

    def __str__(self):
        return "La {} tiene {} y este es su botanico principal: {}".format(self.__class__.__name__, self.aromas, self.botanico_principal)

class Whisky(Bebidas_alcoholicas):
    def __init__(self, gaseosa, graduacion, color, origen, tipo_barril):
        super().__init__(gaseosa, graduacion, color)
        self.__origen = origen
        self.__tipo_barril = tipo_barril

    @property 
    def origen(self):
        return self.__origen

    @origen.setter
    def origen(self, nuevo_origen):
        self.__origen = nuevo_origen

    @property 
    def tipo_barril(self):
        return self.__tipo_barril

    @tipo_barril.setter
    def tipo_barril(self, nuevo_tipo_barril):
        self.__tipo_barril = nuevo_tipo_barril

    def fermentar(self): # funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("El whisky se destila y envejece en barriles de {}.".format(self.tipo_barril))

    def __str__(self):
        return "El {} es de origen {} y se madura en barricas de {}".format(self.__class__.__name__, self.origen, self.tipo_barril)
        
class Ron(Bebidas_alcoholicas):
    def __init__(self, gaseosa, graduacion, color, melaza_o_jugo, envejecimiento):
        super().__init__(gaseosa, graduacion, color)
        self.__melaza_o_jugo = melaza_o_jugo
        self.__envejecimiento = envejecimiento

    @property 
    def melaza_o_jugo(self):
        return self.__melaza_o_jugo 

    @melaza_o_jugo.setter
    def melaza_o_jugo(self, nuevo_melaza_o_jugo):
        self.__melaza_o_jugo = nuevo_melaza_o_jugo

    @property 
    def envejecimiento(self):
        return self.__envejecimiento

    @envejecimiento.setter
    def envejecimiento(self, nuevo_envejecimiento):
        self.__envejecimiento = nuevo_envejecimiento

    def fermentar(self): # funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("El ron se fermenta utilizando {} y luego se envejece durante {}.".format(self.melaza_o_jugo, self.envejecimiento))

    def __str__(self):
        return "El {} esta hecho con {} y es de envejecimiento {}".format(self.__class__.__name__, self.melaza_o_jugo,
                                                                                  self.envejecimiento)
        
class Cerveza(Bebidas_alcoholicas):
    def __init__(self, gaseosa, graduacion, color, tipo_lupulo, nivel_amargor):
        super().__init__(gaseosa, graduacion, color)
        self.__tipo_lupulo = tipo_lupulo
        self.__nivel_amargor = nivel_amargor

    @property 
    def tipo_lupulo(self):
        return self.__tipo_lupulo

    @tipo_lupulo.setter
    def tipo_lupulo(self, nuevo_tipo_lupulo):
        self.__tipo_lupulo = nuevo_tipo_lupulo

    @property 
    def nivel_amargor(self):
        return self.__nivel_amargor

    @nivel_amargor.setter
    def nivel_amargor(self, nuevo_nivel_amargor):
        self.__nivel_amargor = nuevo_nivel_amargor

    def fermentar(self): #  funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("La cerveza se fermenta utilizando {} y tiene un nivel de amargor {}".format(self.tipo_lupulo, self.nivel_amargor))

    def __str__(self):
        return "La {} tiene un lúpulo {}".format(self.__class__.__name__, self.tipo_lupulo)

class Vino(Bebidas_alcoholicas):
    def __init__(self, gaseosa, graduacion, color, variedad_uva, cosecha):
        super().__init__(gaseosa, graduacion, color)
        self.__variedad_uva = variedad_uva
        self.__cosecha = cosecha

    @property 
    def variedad_uva(self):
        return self.__variedad_uva

    @variedad_uva.setter
    def variedad_uva(self, nuevo_variedad_uva):
        self.__variedad_uva = nueva_variedad_uva

    @property 
    def cosecha(self):
        return self.__cosecha

    @cosecha.setter
    def cosecha(self, nueva_cosecha):
        self.__cosecha = nueva_cosecha

    def fermentar(self): # funcion polimorfica,ya que en base al tipo de bebida la fermentación sera una u otra
        print("El vino se fermenta a partir de uvas {} y tiene una cosecha de {}.".format(self.variedad_uva, self.cosecha))

    def __str__(self):
        return "El {} tiene una variedad de uva {} y un color {}".format(self.__class__.__name__, self.variedad_uva, self.color)

# Diccionarios con clave-valor
g1 = Ginebra(gaseosa = "Con gas", graduacion=40,color = "Transparente", aromas = "Enebros, cilantro, cítricos", botanico_principal = "Enebro")

w1 = Whisky(gaseosa = "Sin gas", graduacion = 45, color = "Dorado", origen = "Escocia", tipo_barril = "Barril de roble")

r1 = Ron(gaseosa = "Con gas", graduacion=38, color = "Ambar", melaza_o_jugo = "Melaza", envejecimiento = "Añejo")

c1 = Cerveza(gaseosa = "Con gas", graduacion = 5, color = "Rubia", tipo_lupulo = "Citra", nivel_amargor = "Moderado")

v1 = Vino( gaseosa = "Sin gas", graduacion = 12,color = "Rojo", variedad_uva = "Cabernet Sauvignon", cosecha = 2015)


print(g1)
print(w1)
print(r1)
print(c1)
print(v1)
print(g1.aromas)
print(w1.origen)
print(r1.envejecimiento)
print(c1.nivel_amargor)
print(v1.variedad_uva)
v1.fermentar()
g1.fermentar()
c1.beber()
w1.mezclar()
```
