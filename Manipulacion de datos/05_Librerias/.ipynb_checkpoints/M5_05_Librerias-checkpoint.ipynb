**1) Trabajemos con NumPy**

* Importa numpy y usa un alias para poder utilizar sus recursos
* Crea un array de elementos (del 10 al 100)
* Invertir el array
* Crear un array 3x3 con valores de 0 a 8
* Crear un array identidad de 6x6
* Crear un array con valores al azar con forma 3x3x3
* Crear dos arrays con valores al azar A y B, verificar si son iguales (hay varias opciones, investigar)

```python
import numpy as np
elementos = np.arange(10, 101)
print("Array de elementos del 10 al 100: \n",elementos)
array_invertido = elementos[::-1]
print("Array invertido: \n ", array_invertido)
una_dimension = np.arange(0, 9)
tres_x_tres = una_dimension.reshape(3,3)
print("Array del 0 al 8 en tres dimensiones: \n", tres_x_tres)
matriz_identidad = np.eye(6)
print("Esta es una matriz de identidad: \n", matriz_identidad)
valores_azar = np.random.rand(3,3,3)
print("Array con valores al azar: \n", valores_azar)
A = np.random.rand(1)
B = np.random.rand(1)
print("2 arrays con valores al azar")
print("A:")
print(A)
print("B:")
print(B)
iguales = np.array_equal(A, B)
print("¿Los arrays generado aleatoriamente son iguales?", iguales)
```

**2) Sigamos con NumPy**

* Crear un array de 100 elementos aleatorios comprendidos entre 1 y 1000 (pueden ser números enteros).
* Este array se llamara dólares
* Crear otro array llamado euros
* Realizar la conversión de los precios en dólares del primer array al segundo array en euros. Se puede serguir la siguiente conversión:
    - 1 EUR = 1,14362 USD
    - 1 USD = 0,874413 EUR
* Mostrar el resultado del array de euros pero redondeado a 2 decimales

```python
dolares = np.random.randint(1, 1000, size = 100)
print("Array de dólares: \n", dolares)
tasa_conversion = 0.874413
euros = dolares * tasa_conversion
redondeo = np.round(euros, 2)
print("Array de dólares, transformándolo al valor del euro: \n",  redondeo)
```

```python
import pandas as pd
# abrir el csv en pandas
df =pd.read_csv("05_05_imdb_titulos.csv", delimiter = ",")
df
```

```python
# mostrar los 5 primeros de la lista
df.head()
```

```python
# mostrar el numero toal de registros
num_registros = df.shape[0]
print(num_registros)
```

```python
# ordenar la tabla de mas antigua a mas nueva e imprimir las 5 mas antiguas
df_ordenado = df.sort_values(by = "year", ascending = True)
df_ordenado.head()
```

```python
# mostrar las peliculas con el titulo que contenga la palabra Dracula
pelis_dracula = df.loc[df["title"].str.contains("Dracula", case = False, na = False), :]
num_pelis_dracula = df.shape[0]
print(num_pelis_dracula)
```

```python
#mostrar los 10 titulos mas comunes
top10 = df["title"].value_counts().head(10)
print(top10)
```

```python
# mostrar la primera pelicula mas antigua de Romeo and Juliet
romeo_y_julieta = df.loc[df["title"].str.contains("Romeo y Juliet", case = False, na = False), :]
romeo_y_julieta_orden = romeo_y_julieta.sort_values(by = "year", ascending = False)
pelicula_mas_antigua = romeo_y_julieta_orden.head(1)
print(pelicula_mas_antigua)
```

```python
# listar todas las peliculas que contengan la palabra exorcist
pelis_exorcist = df.loc[df["title"].str.contains("Exorcist", case = False, na = False), :]
pelis_ordenadas = pelis_exorcist.sort_values(by = "year", ascending = False)

if pelis_exorcist.empty:
    print("No hay películas que contengan la palabra Exorcist")
else:
    print(pelis_exorcist)
pelis_exorcist
```

```python
# Mostrar cuantas peliculas fueron hechas en el año 1950
ano_1950 = df["year"] == 1950
condicion_1950 = df.loc[ano_1950, :]
condicion_1950
```

```python
df =pd.read_csv("05_05_imdb_elenco.csv", delimiter = ",")
df
```

```python
# mostrar 5 primeras de la tabla
df.head()
```

```python
# imprimir numero total de registros
num_registros = df.shape[0]
print(num_registros)
```

```python
#Mostrar todos los roles o papeles que hubo en la pelicula "The Godfather". También mostrar el número total de coincidencias
godfather =df.loc[df["title"].str.contains("The Godfather", case = False, na = False), :]
print("El numero total de registros es: {}".format(godfather["character"].count()))
godfather["character"]
```

```python
#Mostrar el elenco completo ordenado por la clasificacion "n" de la pelicula "Dracula" de 1958
titulo = df["title"] == "Dracula"
año = df["year"] == 1958
dracula_1958 = df.loc[titulo & año, ["n", "name"]]
dracula_1958.sort_values(by = "n", ascending = True)
```

```python
# Mostrar cuantos papeles de "Bruce Wayne" han sido hechos en la historia de las peliculas
Bruce_Wayne = df["name"] == "Bruce Wayne"
condicion = df.loc[Bruce_Wayne, :]
cantidad_peliculas_bruce = condicion.shape[0]
print(f"Cantidad de películas en las que ha actuado Bruce Wayne: {cantidad_peliculas_bruce}")
```

```python
# Mostrar cuantos papeles ha hecho "Robert De Niro" en su carrera
robert= df["name"] == "Robert De Niro"
condicion = df.loc[robert, :]
cantidad_peliculas_robert = condicion.shape[0]
print(f"Cantidad de películas en las que ha actuado Robert De Niro: {cantidad_peliculas_robert}")
```

```python
#Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente
charlton = df["name"] == "Charlton Heston"
n = df["n"] == 1
año = (df["year"] >= 1960) & (df["year"] <= 1969)
condicion = df.loc[charlton & n & año, ["character"]]
condicion
```

```python
# Mostrar cuantos papeles para actores hubo en la década de los 50's
decada_50 = (df["year"] >= 1950) & (df["year"] <= 1959)
genero = df["type"] == "actor"
condicion = df.loc[decada_50 & genero, :]
total_papeles = condicion.shape[0]
print("El total de actores que hubo en la decada de 1950 fueron: {}".format(total_papeles))
```

```python
#Mostrar cuantos papeles para actrices hubo en la década de los 50's
decada_50 = (df["year"] >= 1950) & (df["year"] <= 1959)
genero = df["type"] == "actress"
condicion = df.loc[decada_50 & genero, :]
total_papeles = condicion.shape[0]
print("El total de actrices que hubo en la decada de 1950 fueron: {}".format(total_papeles))
```
