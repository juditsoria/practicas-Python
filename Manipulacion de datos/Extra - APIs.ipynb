# Acceso a APIs

Las API (Interfaces de Programación de Aplicaciones) son conjuntos de reglas, protocolos y herramientas que permiten que diferentes aplicaciones, sistemas o componentes se comuniquen entre sí. Las API definen cómo se deben realizar las solicitudes y respuestas entre los sistemas, permitiendo que las aplicaciones interactúen de manera programática y coherente.

### ¿Para qué sirven las APIs?
1. __Integración entre Sistemas__: Las APIs permiten que diferentes sistemas, escritos en diferentes lenguajes de programación y ejecutados en diferentes plataformas, se comuniquen e intercambien datos. Por ejemplo, un sistema de gestión de inventario podría comunicarse con una tienda en línea a través de una API para mantener actualizada la disponibilidad de productos.

2. __Acceso a Funcionalidades de Terceros__: Muchas empresas ofrecen APIs públicas que permiten acceder a sus servicios y datos. Por ejemplo, la API de Google Maps se puede utilizar para incorporar mapas y funciones de geolocalización en una aplicación.

3. __Desarrollo Rápido y Modular__: Las APIs permiten dividir una aplicación en componentes más pequeños y manejables, cada uno con su propia interfaz bien definida. Esto facilita el desarrollo, la prueba y la mantenibilidad del código.

4. __Facilitar la Colaboración__: Al establecer reglas claras para la interacción entre diferentes partes de un sistema, las APIs facilitan la colaboración entre diferentes equipos de desarrolladores. Cada equipo puede trabajar en su propia parte del sistema, siempre y cuando cumpla con la API definida.

5. __Expandir la Distribución de Servicios__: Las empresas pueden ofrecer sus servicios a más clientes y desarrolladores al proporcionar una API. Por ejemplo, una API de procesamiento de pagos permite a los desarrolladores incorporar fácilmente la capacidad de aceptar pagos en línea en sus aplicaciones.

6. __Interoperabilidad__: Las APIs permiten la interoperabilidad entre diferentes sistemas y dispositivos, asegurando que puedan trabajar juntos independientemente de las diferencias en la tecnología subyacente.

### Tipos de APIs:
1. __APIs Web o RESTful (las más utilizadas)__: Utilizan protocolos web como HTTP y suelen ser accesibles a través de Internet. Estas APIs permiten acceder a servicios y datos utilizando métodos HTTP estándar como GET, POST, PUT y DELETE.
2. __APIs de Bibliotecas o Frameworks__: Estas son interfaces ofrecidas por bibliotecas o frameworks de programación para facilitar tareas específicas como el procesamiento de imágenes o la conexión a bases de datos.
3. __APIs de Sistema Operativo__: Proporcionan acceso a funciones del sistema operativo, como la gestión de archivos o la interfaz de usuario.
En resumen, las APIs son fundamentales en la programación moderna, ya que permiten la comunicación, colaboración e integración entre diferentes sistemas, aplicaciones y componentes.

Librería más utilizada en Python: __REQUESTS__ -> https://requests.readthedocs.io/es/latest/




### Métodos HTTP en RESTful
1. __GET__:
    - Acción: Recuperar datos de un recurso.
    - Uso: Solicitar información de un recurso específico o una colección de recursos.
    - Cuerpo de la Solicitud: No suele llevar un cuerpo en la solicitud.
    - Idempotencia: Sí (repetir la solicitud no cambia el estado del recurso).
2. __POST__:
    - Acción: Crear un nuevo recurso.
    - Uso: Enviar datos para crear un nuevo recurso en el servidor.
    - Cuerpo de la Solicitud: Contiene los datos del nuevo recurso a crear.
    - Idempotencia: No (repetir la solicitud puede crear múltiples recursos).
3. __PUT__:
    - Acción: Actualizar o crear un recurso específico.
    - Uso: Actualizar un recurso existente o crear uno nuevo si no existe.
    - Cuerpo de la Solicitud: Contiene los datos completos del recurso a actualizar.
    - Idempotencia: Sí (repetir la solicitud no cambia el estado del recurso tras la primera actualización exitosa).
4. __PATCH__:
    - Acción: Actualizar parcialmente un recurso.
    - Uso: Modificar solo ciertos atributos de un recurso existente.
    - Cuerpo de la Solicitud: Contiene solo los datos que se quieren modificar.
    - Idempotencia: No necesariamente (depende de la implementación y de los datos enviados).
5. __DELETE__:
    - Acción: Eliminar un recurso.
    - Uso: Borrar un recurso específico del servidor.
    - Cuerpo de la Solicitud: No suele llevar un cuerpo en la solicitud.
    - Idempotencia: Sí (repetir la solicitud no cambia el estado, ya que el recurso ya fue eliminado).

En resumen, estos métodos HTTP proporcionan una forma estandarizada de realizar operaciones CRUD (Crear, Leer, Actualizar, Eliminar) en recursos a través de una API RESTful. Permiten una comunicación clara y coherente entre el cliente y el servidor, y su uso adecuado ayuda a mantener la arquitectura limpia y comprensible.

### Códigos de estado

- 200: Todo salió bien y se devolvió el resultado (si lo hubiera).
- 301: el servidor lo redirige a un punto final diferente. Esto puede suceder cuando una empresa cambia los nombres de dominio o se cambia el nombre de un punto final.
- 400: El servidor cree que realizó una solicitud incorrecta. Esto puede suceder cuando no envía los datos correctos, entre otras cosas.
- 401: el servidor cree que no está autenticado. Muchas API requieren credenciales de inicio de sesión, por lo que esto sucede cuando no envía las credenciales correctas para acceder a una API.
- 403: El recurso al que intentas acceder está prohibido: no tienes los permisos adecuados para verlo.
- 404: el recurso al que intentó acceder no se encontró en el servidor.
- 503: El servidor no está listo para manejar la solicitud.

```python
import requests
import folium

# Solicitar la ubicación actual de la ISS
response = requests.get("http://api.open-notify.org/iss-now.json")
data = response.json()

# Extraer las coordenadas de latitud y longitud
latitude = float(data['iss_position']['latitude'])
longitude = float(data['iss_position']['longitude'])

# Crear un mapa centrado en las coordenadas de la ISS con un nivel de zoom de 1 (más zoom más cerca)
iss_map = folium.Map(location=[latitude, longitude], zoom_start=1)

# Agregar un marcador en la ubicación de la ISS
folium.Marker([latitude, longitude], tooltip='ISS Location').add_to(iss_map)

# Mostrar el mapa
iss_map
```

### Otro ejemplo para ver como funcionan los parametros en una API
A cotninuacion vamos a ver un ejemplo utilizando la API pública de JSONPlaceholder (publicaciones de usuarios). Esta API no requiere registro ni clave de API y te permite realizar varias operaciones, incluyendo solicitudes GET con parámetros.

JSONPlaceholder (https://jsonplaceholder.typicode.com/) es una API en línea gratuita que simula una REST API. Puedes usarla para hacer pruebas y prototipos. La API ofrece endpoints para manipular recursos como publicaciones, comentarios, usuarios, álbumes y más.

Aquí tienes una visión general de algunos de los principales endpoints y operaciones que puedes realizar:

- Publicaciones (/posts):

    - GET /posts: Obtiene todas las publicaciones.
    - GET /posts/1: Obtiene la publicación con ID 1.
    - GET /posts/1/comments: Obtiene los comentarios de la publicación con ID 1.
    - POST /posts: Crea una nueva publicación.
    - PUT /posts/1: Actualiza la publicación con ID 1.
    - DELETE /posts/1: Elimina la publicación con ID 1.

- Comentarios (/comments):

    - GET /comments: Obtiene todos los comentarios.
    - GET /comments?postId=1: Obtiene los comentarios de la publicación con ID 1.

- Usuarios (/users):

    - GET /users: Obtiene todos los usuarios.
    - GET /users/1: Obtiene el usuario con ID 1.

- Álbumes (/albums):

    - GET /albums: Obtiene todos los álbumes.
    - GET /albums/1/photos: Obtiene las fotos del álbum con ID 1.

- Tareas Pendientes (/todos):

    - GET /todos: Obtiene todas las tareas pendientes.
    - GET /todos?userId=1: Obtiene las tareas pendientes del usuario con ID 1.

También hay varios parámetros que puedes utilizar para filtrar y modificar las respuestas, como _limit, _page, _sort, y _order.

```python
import requests

def crear_post(url, data):
    """ Crea un nuevo post """
    response = requests.post(url, json=data)
    if response.status_code == 201:
        print("\nPost creado (POST):")
        print(response.json())
    else:
        print("No se pudo crear el post.")
    return response.json()

def obtener_post(url):
    """ Obtiene y muestra la información de un post """
    response = requests.get(url)
    if response.status_code == 200:
        print("Información del post (GET):")
        print(response.json())
    else:
        print("No se pudo obtener el post.")
    return response.json()

def actualizar_post(url, data):
    """ Actualiza la información de un post """
    response = requests.put(url, json=data)
    if response.status_code == 200:
        print("\nPost actualizado (PUT):")
        print(response.json())
    else:
        print("No se pudo actualizar el post.")
    return response.json()

def eliminar_post(url):
    """ Elimina un post """
    response = requests.delete(url)
    if response.status_code == 200:
        print("\nPost eliminado (DELETE)")
    else:
        print("No se pudo eliminar el post.")

# URL base para posts
url_base = "https://jsonplaceholder.typicode.com/posts"
post_id = "1"
url_post = url_base + "/" + post_id

# Datos para un nuevo post
datos_nuevo_post = {
    "title": 'Nuevo Post',
    "body": 'Contenido del nuevo post',
    "userId": 1
}

# Obtener información del post -> GET
post_actual = obtener_post(url_post)

# Crear un nuevo post -> POST
post_creado = crear_post(url_base, datos_nuevo_post)

# Nuevos datos para actualizar el post
nuevos_datos = {
    "title": 'Post Actualizado',
    "body": 'Contenido actualizado del post',
    "userId": 1
}

# Actualizar el post -> PUT
post_actualizado = actualizar_post(url_post, nuevos_datos)

# Eliminar el post -> DELETE
eliminar_post(url_post)
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

params = {
    'userId': 1,   # ID del usuario
    '_limit': 5    # Límite de resultados
}

response = requests.get(url, params=params)

if response.status_code == 200:
    posts_data = response.json()
    print(f"Publicaciones del Usuario {params['userId']} (Límite {params['_limit']}):")
    for post in posts_data:
        print("-", post['title'])
else:
    print("Error en la solicitud:", response.status_code)
```

```python
parametros = {
    "id": 1, 
    "nombre": "Cristian"
}

r = requests.get("https://httpbin.org/get", params = parametros)
print(r.text)
print(r.json())
print(r.url)
```

## Ejemplo sobre SWAPI (Star Wars)
Vamos a combinar APIs con requests, Pandas y Excel

```python
# swapi: API sobre el mundo de Star Wars. No requiere registro

import requests
import pandas as pd

def mostrar_categoria(categoria_url):
    response = requests.get(categoria_url)
    if response.status_code == 200:
        datos = response.json()
        results = datos['results']
        
        # Convertimos la lista de resultados en un DataFrame
        df = pd.DataFrame(results)
        
        # Mostramos las primeras filas del DataFrame
        print(df.head())
        
        return df # Puedes retornar el DataFrame si quieres usarlo más adelante
    else:
        print("Error al obtener datos de la categoría.")
        return None

# Obtenemos información
response = requests.get("https://swapi.dev/api/")
datos = response.json()

# Mostramos información general
print("> Datos:", datos)
print("> Tipo:", type(datos))
print("> Elementos:", len(datos))

opciones = list(datos.keys())
for i, opcion in enumerate(opciones, 1):
    print(f">> {i}. {opcion}")

# Mostramos menú
print("\nBienvenido/a a la biblioteca del Templo Jedi")
opcion_seleccionada = int(input("Seleccione una categoría por número (1-6): "))

if 1 <= opcion_seleccionada <= 6:
    categoria_seleccionada = opciones[opcion_seleccionada - 1]
    print(f"Mostrando datos para la categoría: {categoria_seleccionada}")
    df = mostrar_categoria(datos[categoria_seleccionada])
else:
    print("Opción no válida.")
```

```python
# Solicitamos la información a la API y la exportamos a pandas y posteriormente a excel (las 6 partes)
import requests
import pandas as pd

def obtener_categoria(categoria_url):
    response = requests.get(categoria_url)
    if response.status_code == 200:
        datos = response.json()
        results = datos['results']
        
        # Convertimos la lista de resultados en un DataFrame
        df = pd.DataFrame(results)
        
        # Devolvemos el DataFrame
        return df
    else:
        print("Error al obtener datos de la categoría.")
        return None

# Obtenemos información
response = requests.get("https://swapi.dev/api/")
datos = response.json()

# Crear un objeto ExcelWriter para escribir en un archivo Excel
with pd.ExcelWriter('star_wars_data.xlsx') as writer:
    for categoria, url in datos.items():
        print(f"Obteniendo datos para la categoría: {categoria}")
        df = obtener_categoria(url)
        if df is not None:
            # Escribe el DataFrame en una hoja con el nombre de la categoría
            df.to_excel(writer, sheet_name=categoria)
        else:
            print(f"Error al obtener datos para la categoría: {categoria}")

print("\nDatos exportados exitosamente a 'star_wars_data.xlsx'")
```

## Ejemplo sobre API astros + PIL (pillow)
Vamos a combinar APIs con mostrar una imágen modificada por pantalla

```python
from PIL import Image, ImageDraw, ImageFont
import requests
from IPython.display import display

# API que nos comunica cuántas personas se encuentran actualmente en el espacio
response = requests.get("http://api.open-notify.org/astros.json")
data = response.json()
num = data["number"]
print("Número de personas en el espacio:", num)

# Abrir la imagen
image_path = "space.png"  # Asegúrate de poner la ruta correcta a tu imagen
image = Image.open(image_path)

# Crear objeto ImageDraw para dibujar en la imagen
draw = ImageDraw.Draw(image)

# Coordenadas (x, y) del centro de la imagen
x, y = image.size[0] / 2, image.size[1] / 2

# Fuente y tamaño
font_size = 40
font = ImageFont.truetype("arial.ttf", font_size)

# Calcular la posición para centrar el texto en la imagen
text = str(num)
text_box = draw.textbbox((x, y), text, font=font)
text_width = text_box[2] - text_box[0]
text_height = text_box[3] - text_box[1]
x -= text_width / 2
y -= text_height / 2
# En este punto x e y apuntan al centro

# Escribir el número en la imagen
draw.text((x+170, y-70), text, font=font, fill="white")

# Mostrar la imagen en Jupyter Notebook
display(image)
```
