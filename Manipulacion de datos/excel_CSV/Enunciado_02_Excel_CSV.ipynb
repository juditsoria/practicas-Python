**4) Continuamos con Excel. Crea una nueva columna que sea el sumatorio de todas las medallas conseguidas por cada uno de los paises**

````python
wb = openpyxl.load_workbook("02_Excel_data.xlsx") 
sheet = wb["olimpiadas"]
for row in range(2, sheet.max_row + 1):
    oro = sheet.cell(row=row, column=2).value  # Columna B: Oro
    plata = sheet.cell(row=row, column=3).value  # Columna C: Plata
    bronce = sheet.cell(row=row, column=4).value  # Columna D: Bronce
    
    # Calcular el sumatorio de las medallas, asegurándose de que los valores no sean None
    total_medallas = (oro if oro else 0) + (plata if plata else 0) + (bronce if bronce else 0)
    
    # Escribir el total en la nueva columna E
    sheet.cell(row=row, column=5, value=total_medallas)  # Columna E

# Guardar el archivo con los cambios
wb.save('02_Excel_data.xlsx')
for fila in hoja_olimpiadas.iter_rows(values_only = True):
    print(", ".join(fila))
````

**5) Continuamos con Excel. Crear un gráfico de barras (openpyxl.chart.BarChart()) lo más parecido al que se muestra en la imagen:**



````python
from openpyxl import Workbook
from openpyxl.chart import BarChart, Reference, Series
from openpyxl.chart.axis import TextAxis

# Datos de ejemplo
data = {
    'País': ['USA', 'China', 'UK', 'Russia', 'South Korea', 'Germany'],
    'Medallas': [46, 38, 29, 22, 13, 11]  # Totales de medallas por país
}

# Crear un libro de trabajo y una hoja
wb = Workbook()
hoja_olimpiadas = wb.active
hoja_olimpiadas.title = 'Medallas'

# Agregar los encabezados
hoja_olimpiadas.append(['País', 'Medallas'])

# Agregar los datos a la hoja
for i in range(len(data['País'])):
    hoja_olimpiadas.append([data['País'][i], data['Medallas'][i]])

# Crear el gráfico de barras
grafico = BarChart()
grafico.type = "col"
grafico.style = 10
grafico.title = "Medallas Olímpicas"
grafico.y_axis.title = 'Total de Medallas'

# Asignar colores personalizados por país
colores = {
    'USA': '0000FF',  # Azul
    'China': 'FFA500',  # Naranja
    'UK': '808080',  # Gris
    'Russia': 'FFFF00',  # Amarillo
    'South Korea': '0000FF',  # Azul
    'Germany': '008000'  # Verde
}

# Crear una serie para cada país con su color
for i, pais in enumerate(data['País'], start=2):  
    data_ref = Reference(hoja_olimpiadas, min_col=2, min_row=i, max_col=2, max_row=i)
    serie = Series(data_ref, title=pais)  
    serie.graphicalProperties.solidFill = colores.get(pais, 'FFFFFF')  # Asignar color en base a los colores arriba indicados
    grafico.append(serie)

# Configurar las categorías (nombres de los países en el eje X)
categories_reference = Reference(hoja_olimpiadas, min_col=1, min_row=2, max_row=7)
grafico.set_categories(categories_reference)

# Configurar el espacio entre barras
grafico.gapWidth = 300  # Aumentar espacio entre barras
grafico.overlap = -100  # Desactivar agrupación de barras

# Configurar el eje X para que se vean los nombres de los países
grafico.x_axis = TextAxis()  # Aseguramos que el eje X sea de tipo texto
grafico.x_axis.title = 'País'  # Título del eje X
grafico.x_axis.tickLblPos = "low"  # Posiciona las etiquetas cerca del eje


# Agregar el gráfico a la hoja
hoja_olimpiadas.add_chart(grafico, "E5")

# Guardar el archivo
wb.save('grafica.xlsx')


````

**6) Observa los datos del csv y realiza las siguientes tareas:**

* Tenemos 9 columnas, las 8 primeras contienen datos numéricos con los cuales podemos trabajar, crea una lista para cada una de estas columnas (mpg, cylinders, etc.). No es necesario crear lista para name
* Recorre los datos del csv adecuadamente y almacena los datos de cada columna en cada una de las listas que has creado anteriormente.
* Comprobar que se haya guardado en las listas la información correspondiente (con prints)

````python
import csv

# Crear listas para cada columna
mpg = []
cylinders = []
displacement = []
horsepower = []
weight = []
acceleration = []
model_year = []
origin = []

# Leer el archivo CSV
with open('02_CSV_data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Saltamos la cabecera si tiene una
    
    # Recorremos cada fila y guardamos los datos en las listas correspondientes
    for row in reader:
        mpg.append(float(row[0]))  # Convertir el valor a float si es necesario
        cylinders.append(int(row[1]))  # Convertir a int
        displacement.append(float(row[2]))
        horsepower.append(float(row[3]))
        weight.append(float(row[4]))
        acceleration.append(float(row[5]))
        model_year.append(int(row[6]))
        origin.append(int(row[7]))

# Imprimir los datos para comprobar que se han guardado correctamente
print("MPG:", mpg[:5])  # Muestra los primeros 5 elementos para verificar
print("Cylinders:", cylinders[:5])
print("Displacement:", displacement[:5])
print("Horsepower:", horsepower[:5])
print("Weight:", weight[:5])
print("Acceleration:", acceleration[:5])
print("Model Year:", model_year[:5])
print("Origin:", origin[:5])
````

**7) Vamos a trabajar con los números de nuestras listas por lo que nos sobra el primer elemento de cada lista, el cual contiene el nombre de cada una de las columnas.**

````python
import csv

# Leer el archivo CSV
with open('02_CSV_data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    # Convertimos el CSV en una lista de listas (cada fila como una sublista)
    data = list(csv_reader)

# Eliminar el primer elemento (el nombre de las columnas)
data.pop(0)

# Convertir los valores a flotantes, manejando posibles errores
for row in data:
    try:
        # Convertir cada elemento de la fila a float
        row = [float(value) if value.replace('.', '', 1).isdigit() else 0.0 for value in row]
        print(row)  # Mostramos la fila convertida a float
        print(type(row[0]))  # Verificamos que el primer valor sea un float
    except ValueError as e:
        print(f"Error al convertir la fila {row}: {e}")


````

**8) Para terminar vamos a mostrar algunos resultados. Para ello utilizaremos la librería statistics.**

* Muestra para cada una de las listas, cuantas observaciones tiene (cuántos datos tiene).
* Mostrar el mínimo y el máximo de cada lista
* Mostrar la media, mediana y desviación estándar de cada lista (utilizando statistics)

````python
import csv
import statistics

# Leer el archivo CSV
with open('02_CSV_data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    # Convertimos el CSV en una lista de listas (cada fila como una sublista)
    data = list(csv_reader)

# Eliminar el primer elemento (el nombre de las columnas)
data.pop(0)

# Convertir los valores a flotantes, manejando posibles errores
for row in data:
    row = [float(value) if value.replace('.', '', 1).isdigit() else 0.0 for value in row]

    # Mostrar cuántas observaciones tiene cada lista
    print(f"Cantidad de observaciones: {len(row)}")
    
    # Mostrar el mínimo y máximo de cada lista
    print(f"Mínimo: {min(row)}")
    print(f"Máximo: {max(row)}")
    
    # Calcular y mostrar la media, mediana y desviación estándar
    try:
        mean = statistics.mean(row)
        median = statistics.median(row)
        stdev = statistics.stdev(row)
        
        print(f"Media: {mean}")
        print(f"Mediana: {median}")
        print(f"Desviación estándar: {stdev}")
    except statistics.StatisticsError as e:
        print(f"Error al calcular estadísticas para la fila {row}: {e}")
    
    print("-" * 40)  # Separador entre filas
````
