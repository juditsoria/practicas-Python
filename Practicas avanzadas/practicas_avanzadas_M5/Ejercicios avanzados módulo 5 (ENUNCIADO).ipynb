````python
import json
import pprint
import sqlite3 as sq
from sqlite3 import OperationalError
import os


def abrir_fichero_json():
    ''' abrir_fichero_json sirve para abrir el fichero json y mostrar la informacion por pantalla. '''
    with open("incidencias.json", "r") as file:
        data = json.load(file)
        print(f"El tipo de datos en el archivo incidencias.json es: {type(data)}")
    pp = pprint.PrettyPrinter(indent=4) # se imprime el objeto json mas estructurado
    pp.pprint(data)
    


def añadir_incidencia():
    """Permite añadir una nueva incidencia al fichero JSON."""
    nueva_incidencia = {
        "id": input("Ingrese ID de la incidencia: "),
        "tipo": input("Ingrese el tipo de incidencia: "),
        "aerolinea": input("Ingrese el nombre de la aerolínea: "),
        "numero": input("Ingrese el número de vuelo: "),
        "origen": input("Ingrese la ciudad de origen: "),
        "destino": input("Ingrese la ciudad de destino: "),
        "hora_salida": input("Ingrese la hora de salida (formato: YYYY-MM-DDTHH:MM:SS): "),
        "hora_llegada": input("Ingrese la hora de llegada (formato: YYYY-MM-DDTHH:MM:SS): "),
        "razon": input("Ingrese la causa de la incidencia: ")
    }
    with open("incidencias.json", "r+") as file: # modo lectura y escritura
        try:
            data = json.load(file)  # transformar de json a dict 
        except json.JSONDecodeError:
            data = {"incidencias": [] }
        data["incidencias"]. append(nueva_incidencia)
        file.seek(0) # mueve el puntero al inicio del archivo
        json.dump(data, file, indent=4)   # transforma a json
        print("La incidencia ha sido creada con éxito.")
        pp = pprint.PrettyPrinter(indent=4)
        pp.pprint(nueva_incidencia)

def crear_BBDD():
    ''' crea la BBDD y define la estructura de la tabla incidencias '''
    try:
        conexion = sq.connect("incidenciasVuelos.db")
        cursor = conexion.cursor()
        cursor.execute('''CREATE TABLE incidencias (id VARCHAR PRIMARY KEY,
                        Tipo VARCHAR,
                        aerolinea VARCHAR,
                        vuelo VARCHAR,
                        origen VARCHAR,
                        destino VARCHAR,
                        horaSalida VARCHAR,
                        horaLlegada VARCHAR,
                        causa VARCHAR)''')
        print("Base de datos creada con éxito.")
        conexion.commit()
        conexion.close()
    except OperationalError as e:
        print("La tabla de incidencias ya existe en la Base de Datos.")

def limpiar_incidencia(incidencia):
    ''' crear una estructura predeterminada para incidencias válidas '''
    incidencia_valida = {
        "id": incidencia.get("id", "desconocido"),
        "tipo": incidencia.get("tipo", "desconocido"),
        "aerolinea": {
            "nombre": incidencia.get("aerolinea", "desconocido"),
            "codigo": "desconocido"
        },
        "vuelo": {
            "numero": incidencia.get("numero", "desconocido"),
            "origen": incidencia.get("origen", "desconocido"),
            "destino": incidencia.get("Destino", "desconocido"),
            "hora_salida": incidencia.get("hora_salida", "00:00"),
            "hora_llegada": incidencia.get("hora_llegada", "00:00")
        },
        "razon": {
            "tipo": "desconocido",
            "descripcion": incidencia.get("razon", "No especificada")
        }
    }
    return incidencia_valida


def actualizar_BBDD():
    ''' Actualiza la BBDD con los nuevos datos que haya dentro del fichero json '''
    try:
        with open("incidencias.json", "r") as file:
            data = json.load(file)
            incidencias = data.get("incidencias", [])
    except FileNotFoundError:
        print("El archivo json no se ha encontrado")
    except json.JSONDecodeError:
        print("Error al leer el archivo JSON.")
        return
    conexion = sq.connect("incidenciasVuelos.db")
    cursor = conexion.cursor()
    for incidencia in incidencias:
        if not (isinstance(incidencia.get("aerolinea"), dict) and
                isinstance(incidencia.get("vuelo"), dict) and
                isinstance(incidencia.get("razon"), dict)):
            print(f"Corrigiendo incidencia inválida: {incidencia}")
            incidencia = limpiar_incidencia(incidencia)
            
        try:
            cursor.execute('''INSERT OR REPLACE INTO incidencias  
                            (id, Tipo, aerolinea, vuelo, origen, destino, horaSalida, horaLlegada, causa)
                            VALUES (:id, :tipo, :aerolinea, :vuelo, :origen, :destino, :hora_salida, :hora_llegada, :razon)
                            ''', {
                                "id": incidencia["id"],
                                "tipo": incidencia["tipo"],
                                "aerolinea": incidencia["aerolinea"]["nombre"],
                                "vuelo": incidencia["vuelo"]["numero"],
                                "origen": incidencia["vuelo"]["origen"],
                                "destino": incidencia["vuelo"]["destino"],
                                "hora_salida": incidencia["vuelo"]["hora_salida"],
                                "hora_llegada": incidencia["vuelo"]["hora_llegada"],
                                "razon": incidencia["razon"]["descripcion"]
                            })
        except sq.IntegrityError as e:
            print(f"Error de integridad con la incidencia {incidencia['id']}: {e}")

    conexion.commit()
    conexion.close()
    print("Base de datos actualizada con éxito.")

def visualizar_incidencias():
    ''' muestra todas las incidencias que hay dentro de la BBDD '''
    try:
        conexion = sq.connect("incidenciasVuelos.db")
        cursor = conexion.cursor()
        cursor.execute("SELECT * FROM incidencias")
        incidencias = cursor.fetchall()
        claves = ['id', 'tipo', 'aerolinea', 'vuelo', 'origen', 'destino', 'hora_salida', 'hora_llegada', 'causa']
        if incidencias:
            for incidencia in incidencias:
                for i, valor in enumerate(incidencia):
                    print(f"{claves[i]}: {valor}")
        else:
            print("No hay incidencias para mostrar")
        
        conexion.close()

    except sq.Error as e:
        print(f"Error al acceder a la base de datos: {e}")

def borrar_BBDD():
    '''Eliminar la BBDD '''
    try:
        ruta = "incidenciasVuelos.db"
        if os.path.exists(ruta):
            os.remove(ruta)
            print("La BBDD se ha eliminado con éxito")
        else:
            print("La BBDD no existe")
        
    except Exception as e:
        print(f"Error al eliminar la BBDD {e}")
def menu():
    """Muestra el menú de opciones y permite al usuario seleccionar una."""
    while True:
        print("Menú de opciones: \n 1. Ver fichero JSON de incidencias")
        print("2. Añadir incidencia al fichero JSON de incidencias")
        print("3. Crear base de datos de incidencias (solo estructura)")
        print("4. Actualizar contenido base de datos de incidencias")
        print("5. Ver contenido base de datos de incidencias")
        print("6. Eliminar base de datos de incidencias")
        print("7. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            abrir_fichero_json()
        elif opcion == "2":
            añadir_incidencia()
        elif opcion == "3":
            crear_BBDD()
        elif opcion == "4":
            actualizar_BBDD()
        elif opcion == "5":
            visualizar_incidencias()
        elif opcion == "6":
            borrar_BBDD()
        elif opcion == "7":
            print("Saliendo del programa...")
            break 

if __name__ == "__main__":
    menu()

````

````python
import pandas as pd
df = pd.read_json("log_tren.json")
df
def abrir_json():
    ''' Abrir el archivo json obtener id, type, level, category, description y timestamp de los objetos
    que estan dentro de la lista dentro del objeto log del archivo json, crear una nueva columna con esos nombres y
    meter dentro los datos del json para poder manipularlo todo después de manera mas eficiente. 
    '''
    df = pd.read_json("log_tren.json")  
    df["type"] = df["log"].apply(lambda x: x["type"]) 
    df ["level"] = df["log"].apply(lambda x: x["level"])
    df ["category"] = df["log"].apply(lambda x: x["category"])
    df ["description"] = df["log"].apply(lambda x: x["description"])
    df ["timestamp"] = df["log"].apply(lambda x: x["timestamp"])
    df ["id"] = df["log"].apply(lambda x: x["id"])
    return df
def mostrar_num_errores():
    '''Mostrar el número total de errores que hay almacenados en el archivo json. 
    '''
    df = abrir_json()
    total_errores = df.shape[0]
    print(f"El total de errores es: {total_errores}")
    
def tipo_errores():
    ''' Mostrar los errores que hay clasificados por tipo (type) que hay y la cantidad de cada uno de ellos.
    '''
    df = abrir_json()
    error_tipo = df["type"].value_counts()
    print("El numero de errores por tipo es: ")
    for tipo, cantidad in error_tipo.items():
        print(f"- {tipo}: {cantidad}")

def errores_nivel():
     ''' Mostrar los errores que hay clasificados por nivel (level) que hay y la cantidad de cada uno de ellos.
    '''
    df = abrir_json()
    error_nivel = df["level"].value_counts()
    print("Numero de errores por nivel:")
    for tipo, cantidad in error_nivel.items():
        print(f"- {tipo}: {cantidad}")

def errores_categoria():
    ''' Mostrar los errores que hay clasificados por categoría (category) que hay y la cantidad de cada uno de ellos.
    '''
    df = abrir_json()
    errores_categoria = df["category"].value_counts()
    print("Numero de errores por categoria:")
    for tipo, cantidad in errores_categoria.items():
        print(f"- {tipo}: {cantidad}")

def palabra_repetida():
    ''' Mostrar las 10 palabras más repetidas en las descripciones (description).
    '''
    df = abrir_json()
    descripciones =" ".join(df["description"]) # junta todas las descriptiones en un str
    palabras = descripciones.split() # forma una lista y mete cada palabra dentro de ella
    palabra_repetida = pd.Series(palabras).value_counts()
    print("Las palabras más repetidas en las descripciones son:")
    print(palabra_repetida.head(10))

def errores_dia():
    ''' Mostrar los errores ordenados por fecha en orden ascendente. 
    '''
    df = abrir_json()
    errores_dia = df["timestamp"].sort_values(ascending = True)
    print(f"Las fechas de los errores ordenadas ascendentemente: \n {errores_dia}")

def buscar_errores():
    ''' Buscador de errores, mediante un input puedes elegur buscar tu error mediante id, tipo de error, nivel de error,
    categoría, descripción y fecha. 
    '''
    print("Criterios de búsqueda disponibles:")
    print("1. Buscar por id")
    print("2. Buscar por tipo de error")
    print("3. Buscar por nivel de error")
    print("4. Buscar por categoría")
    print("5. Buscar por descripción")
    print("6. Buscar por fecha")
    
    opcion = input("Selecciona el número del criterio de búsqueda (1-6): ")
    df = abrir_json()
    if opcion == "1":
        id = input("Introduce el id del error que quieras buscar (LOG0005):")
        resultados = df[df["id"].str.contains(id, case = False, na = False)]
        for index, row in resultados.iterrows():
            print(f"ID: {row["id"]}")
            print(f"Tipo: {row["type"]}")
            print(f"Nivel: {row["level"]}")
            print(f"Categoría: {row["category"]}")
            print(f"Descripción: {row["description"]}")
            print(f"Fecha: {row["timestamp"]}")
    elif opcion == "2":
        tipo = input("Introduce el tipo de error a buscar (ejemplo: 'error', 'warning'): ")
        resultados = df[df["type"].str.contains(tipo, case=False, na=False)]  
    elif opcion == "3":
        nivel = input("Introduce el nivel de error a buscar (ejemplo: 'critical', 'warning'): ")
        resultados = df[df["level"].str.contains(nivel, case=False, na=False)]  
    elif opcion == "4":
        categoria = input("Introduce la categoría de error a buscar: ")
        resultados = df[df["category"].str.contains(categoria, case=False, na=False)] 
    elif opcion == "5":
        descripcion = input("Introduce la palabra o frase a buscar en la descripción: ")
        resultados = df[df["log"].apply(lambda x: descripcion.lower() in x["description"].lower())] 
    elif opcion == "6":
        fecha = input("Introduce la fecha a buscar (formato: YYYY-MM-DD): ")
        resultados = df[df["timestamp"].str.contains(fecha, na=False)]  
    else:
        print("Opción no válida.")
        return
    
    if not resultados.empty:
        print(f"Se han encontrado {len(resultados)} resultados:")
    else:
        print("No se han encontrado resultados que coincidan con los criterios proporcionados.")


def menu():
    ''' Mostrar el menú principal del programa, y que el usuario pueda escoger que quiere hacer mediante un input.
    '''
    print("Seleccione una opción: \n 1. Número total de errores \n 2. Número de errores por tipo (type)")
    print(" 3. Número de errores por nivel (level) \n 4. Número de errores por categoria (category)")
    print("5. Palabras más repetidas en la descripción \n 6. Errores por día \n 7. Búsqueda de errores \n 8. Salir")
    while True: 
        opcion = input("Seleccione una opcion correcta(n)")
        if opcion == "1":
            mostrar_num_errores()
        elif opcion == "2":
            tipo_errores()
        elif opcion == "3":
            errores_nivel()
        elif opcion == "4":
            errores_categoria()
        elif opcion == "5":
            palabra_repetida()
        elif opcion == "6":
            errores_dia()
        elif opcion == "7":
            buscar_errores()
        elif opcion == "8":
            print("Saliendo del programa...")
            break
        

if __name__ == "__main__":
    menu()
````
